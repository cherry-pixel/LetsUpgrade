# Loading the Dataset

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

origdata=pd.read_csv('general_data.csv')
origdata.head(11)
```

```
origdata.columns
```

#Checking Null values and removing them


```
null=origdata.isnull()
null
```

```
origdata.dropna();
origdata
```

# Checking Duplicate values and removing them

```
dupli=origdata.duplicated()
dupli
```

```
origdata.drop_duplicates()
origdata
```

# Measure of Central Tendency 

```
origdata[['MonthlyIncome','YearsAtCompany','DistanceFromHome']].describe()
```

```
origdata[['MonthlyIncome','YearsAtCompany','DistanceFromHome']].median()
```

```
origdata[['MonthlyIncome','YearsAtCompany','DistanceFromHome']].mode()
```

```
origdata[['MonthlyIncome','YearsAtCompany','DistanceFromHome']].var()
```

# Skewness

```
origdata[['MonthlyIncome','YearsAtCompany','DistanceFromHome']].skew()
```

# Kurtosis

```
origdata[['MonthlyIncome','YearsAtCompany','DistanceFromHome']].kurt()
```

- MonthlyIncome, YearsAtCompany = Leptokurtic  (Positive)
- Platyokurtic = DistanceFromHome (Negative)
